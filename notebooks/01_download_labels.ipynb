```python
import os
import pandas as pd
from nilearn import datasets
```

```python
adhd = datasets.fetch_adhd(n_subjects=40)

print("Dataset loaded")
print(adhd.keys())
```

```python
print(type(adhd.phenotypic))
print(adhd.phenotypic.head())
print(adhd.phenotypic.columns.tolist())
```

```python
print("Number of scans:", len(adhd.func))
```

```python
adhd.phenotypic.columns.tolist()
```

```python
pheno = adhd.phenotypic.copy()

participants = pd.DataFrame({
    "subject_id": pheno["Subject"],
    "age": pheno["age"],
    "sex": pheno["sex"],
    "site": pheno["site"],
    "label": pheno["adhd"]   # 1 = ADHD, 0 = Control
})

participants.head()
```

```python
print(participants["label"].value_counts())
```

```python
os.makedirs("data/processed", exist_ok=True)

participants.to_csv(
    "data/processed/participants.csv",
    index=False
)

print("Saved participants.csv")
```
